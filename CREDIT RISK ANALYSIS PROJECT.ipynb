```python
missing_percent = app_data.isnull().mean() * 100
columns_to_drop = missing_percent[missing_percent > 40].index
app_data.drop(columns=columns_to_drop, axis=1, inplace=True)

print(f"Removed {len(columns_to_drop)} columns with >40% missing values.")
```

```python
numeric_cols = app_data.select_dtypes(include=['float64', 'int64']).columns
app_data[numeric_cols] = app_data[numeric_cols].fillna(app_data[numeric_cols].median())

categorical_cols = app_data.select_dtypes(include='object').columns
app_data[categorical_cols] = app_data[categorical_cols].fillna(app_data[categorical_cols].mode().iloc[0])

print("Missing values filled: numeric → median, categorical → mode.")
```

# Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Numeric Features

```python
numeric_features = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'DAYS_BIRTH', 'DAYS_EMPLOYED']

for col in numeric_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(app_data[col], kde=True, bins=40, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

#### Categorical Features

```python
categorical_features = ['NAME_CONTRACT_TYPE', 'CODE_GENDER', 'NAME_EDUCATION_TYPE']

for col in categorical_features:
    plt.figure(figsize=(12, 10))
    sns.countplot(data=app_data, x=col, color='salmon')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

# Bivariate Analysis

#### Numeric vs Target

```python
for col in numeric_features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=app_data, x='TARGET', y=col, palette='Set2')
    plt.title(f'{col} vs TARGET')
    plt.xlabel('TARGET (0: Non-defaulter, 1: Defaulter)')
    plt.ylabel(col)
    plt.show()
```

#### Categorical vs Target

```python
for col in categorical_features:
    plt.figure(figsize=(12, 10))
    sns.countplot(data=app_data, x=col, hue='TARGET', palette='pastel')
    plt.title(f'{col} by TARGET')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

#### Outlier Detection

```python
for col in ['AMT_INCOME_TOTAL', 'AMT_CREDIT']:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=app_data, x=col)
    plt.title(f'Outliers in {col}')
    plt.show()
```

#### Target Imbalance Analysis

```python
target_counts = app_data['TARGET'].value_counts(normalize=True) * 100
target_counts.plot(kind='bar')
plt.title('Target Variable Distribution (%)')
plt.ylabel('Percentage')

print(target_counts)
```

#### Segmented Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt


cat_features = ['NAME_CONTRACT_TYPE', 'CODE_GENDER', 'NAME_EDUCATION_TYPE', 
                'NAME_FAMILY_STATUS', 'NAME_INCOME_TYPE', 'OCCUPATION_TYPE']


for col in cat_features:
    plt.figure(figsize=(7, 4))
    sns.countplot(data=app_data, x=col, hue='TARGET', palette='Set2')
    plt.title(f'Default Rate by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=30, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```
